```python
import pandas as pd
import numpy as np
import time
```

```python
%%time
bio_chim = pd.read_csv('биологическая химия.csv', sep = ';',encoding ='ANSI')
fito = pd.read_csv('фитобиотехнологии.csv', sep = ';',encoding ='ANSI')
neuro = pd.read_csv('нейронауки.csv', sep = ';',encoding ='ANSI')
eco = pd.read_csv('Экология человека.csv', sep = ';',encoding ='ANSI')
```

```python
%%time
bio_chim = bio_chim.drop(['№','Unnamed: 4'], axis = 1)
bio_chim = bio_chim.drop([27,28,29],axis = 0)
```

```python
fito = fito.drop(['№'], axis = 1)
fito = fito.drop([27,28,29,30],axis = 0)
```

```python
neuro = neuro.drop(['№','Unnamed: 5'], axis = 1)
neuro = neuro.drop([27,28,29],axis = 0)
```

```python
eco = eco.drop(['№'], axis = 1)
eco = eco.drop([27,28,29,30],axis = 0)
```

```python
dict_predm = {}
def create_dict(data_frame):
    group = []
    global dict_predm
    for i in list(data_frame.keys()[1:]):
        ind = i.split('\n')
        if '' in ind:
            del ind[ind.index('')]
        group.append(ind[1:])
    for i in range(1,len(group)+1):
        list_name = list(data_frame[data_frame.iloc[:,i] == 1].iloc[:,0])
        for name in list_name:
            if name.split()[0] not in dict_predm.keys():
                dict_predm[name.split()[0]] = set()
            for pred in group[i - 1]:
                dict_predm[name.split()[0]].add(pred.strip(' ').lower())
    
            
            
```

```python
create_dict(bio_chim)
create_dict(fito)
create_dict(neuro)
create_dict(eco)
dict_predm['Мирошников']
dict_predm.keys()
```
